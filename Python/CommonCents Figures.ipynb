```python
from matplotlib import pyplot
%matplotlib inline
```

```python
Px   = [6,5,4,3,2,1,0]
Qdx  = [0,20000,40000,60000,80000,100000,120000]
Qsx0 = [120000,100000,80000,60000,40000,20000,0]
Qsx1 = [140000,120000,100000,80000,60000,40000,20000]
```

```python
pyplot.plot(Qdx, Px,color='#e4e4cb',marker='o',linewidth=6,label='Consumer')
pyplot.plot(Qsx0,Px,color='#f5f5dc',marker='s',linewidth=6,label='Producer')
pyplot.xlabel('Quantity of Commodity X')
pyplot.ylabel('Price in dollars')
pyplot.legend(loc=7,fontsize='medium')
pyplot.grid(True)
pyplot.savefig('c:/Users/Conrad/Pictures/Before Government Intervention',dpi=100,size=(8,4.5))
```
