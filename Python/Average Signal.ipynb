```python
def average_signal(ts1,f,ts2):
    ave_times  = []
    ave_values = []
    for i in range(len(ts2)-1):
        ave_times.append(( ts2[i+1] + ts2[i] ) / 2.0)
        target_indices = np.logical_and(ts1 >= ts2[i],ts1 <= ts2[i+1])
        ave_values.append(np.average(f[target_indices]))
    ave_times  = np.array(ave_times)
    ave_values = np.array(ave_values)     
    
    return ave_times, ave_values
```
